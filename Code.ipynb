```python
def feature_selection(forward_fs:bool, floating_fs:bool, X_train_nofs, y_train):
    from sklearn.ensemble import RandomForestClassifier
    from mlxtend.feature_selection import SequentialFeatureSelector as SFS
    model_fs = RandomForestClassifier(random_state=0) 
    
    sfs = SFS(model_fs, 
           k_features="best", 
           forward=forward_fs, 
           floating=floating_fs, 
           verbose=0,
           scoring="f1_micro",
           n_jobs=-1,
           cv=2)
    sfs1 = sfs.fit(X_train_nofs, y_train)
    return sfs1
```

```python
def obj_function(type_of, X_trva_st, y_trva_st, fold_trva_of, model_of, n_clusters_of):
    from sklearn.metrics import f1_score
    from sklearn.metrics import accuracy_score
    from math import sqrt
    import numpy as np

    X_trva_st = X_trva_st.reset_index(drop=True)
    y_trva_st = y_trva_st.reset_index(drop=True)

    X_trva_f= X_trva_st.copy()
    y_trva_f= y_trva_st.copy()
    

    # HPO - train on raw(3-fold train) and validate on raw(1-fold random)
    if type_of=="rr":
        k_f=4
        i_f=0



        from sklearn.model_selection import KFold
        kf = KFold(n_splits=k_f, shuffle=True, random_state=40)

        for f, (train_index_of, test_index_of) in enumerate(kf.split(X_trva_f)):
            X_trva_f.loc[test_index_of,'kfold']=int(f)
            y_trva_f.loc[test_index_of,'kfold']=int(f)

        fold_trva_k = np.array([0, 1, 2, 3])

        k_acc_of=np.zeros((1,k_f))
      

        for fold_va in fold_trva_k:

            fold_tr = fold_trva_k[fold_trva_k != fold_va]

            X_train_of=X_trva_f[X_trva_f['kfold'].isin(fold_tr)].drop(['kfold'],axis=1)
            y_train_of=y_trva_f[y_trva_f['kfold'].isin(fold_tr)].drop(['kfold'],axis=1).values
            X_validation_of=X_trva_f[X_trva_f['kfold']==fold_va].drop(['kfold'],axis=1) 
            y_validation_of=y_trva_f[y_trva_f['kfold']==fold_va].drop(['kfold'],axis=1).values

            model_of.fit(X_train_of, y_train_of.ravel())
            y_val_pred_of = model_of.predict(X_validation_of)

            k_acc_of[0,i_f]=f1_score(y_validation_of, y_val_pred_of, average='micro')

            i_f=i_f+1

    return (np.mean(k_acc_of))
```

```python
import optuna
import xgboost as xgb
from sklearn.metrics import mean_absolute_percentage_error


# Define the objective function for Optuna to optimize
def objective_xgb(trial):
    params_xgb = {
            'max_depth': trial.suggest_int('max_depth', 3, 150),
            'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.5),
            'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
            'subsample': trial.suggest_float('subsample', 0.01, 1),
            'gamma': trial.suggest_float('gamma', 0, 1),
            'colsample_bytree': trial.suggest_float('colsample_bytree', 0.01, 1),
            'reg_alpha': trial.suggest_float('reg_alpha', 0, 10),
            'reg_lambda': trial.suggest_float('reg_lambda', 0, 10),
            'min_child_weight': trial.suggest_int('min_child_weight', 1, 5)
    }
    

    model_hpo = xgb.XGBClassifier(**params_xgb, random_state=40, n_jobs= -1)
    va_results=obj_function("rr", X_train_d, y_train_d, 0, model_hpo, 10)
 
    return va_results



# Set up the sampler and pruner
sampler = optuna.samplers.TPESampler()

# Create the Optuna study and run the optimization
study_xgb = optuna.create_study(direction='maximize', sampler=sampler)
study_xgb.optimize(objective_xgb, n_trials=50)

# Print the best hyperparameters and corresponding metric
print(f'best trial: {study_xgb.best_trial.number} for fold ')
print(f'xgboost-mape: {study_xgb.best_trial.value:.4f}')
for key, value in study_xgb.best_trial.params.items():
    print(f'  {key}: {value}')
best_params_xgb = study_xgb.best_params
```

```python
best_model = xgb.XGBClassifier(**best_params_xgb, random_state=40,n_jobs= -1)
X_test_d=X_test[list(fsn)].reset_index(drop=True)
y_test_d=y_test.reset_index(drop=True)
from sklearn.metrics import f1_score

best_model.fit(X_train_d ,y_train_d )

# Generate metrics for Train using raw(Train) on best_params 
y_te_pred = best_model.predict(X_test_d)
f1_score(y_test_d, y_te_pred, average='micro')
```
